```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive, files

# Load the netflix_title.csv from my Google Drive
drive.mount('/content/gdrive')

# The following lines prevent some errors while loading data, delete them from the code and check for yourselves:
# - encoding="latin-1"
# - delimiter=","
df=pd.read_csv('gdrive/My Drive/csv/netflix_titles.csv', delimiter=",", encoding="latin-1", parse_dates=["date_added"], index_col=["show_id"])
# Dataset is ranked according to "show_id" column.

# Display the content of retrieved dataframe - df
df
```

```python
df.dtypes
```

```python
# Data Cleaning
df["date_added"] = df["date_added"].dt.year
df["date_added"].unique()
```

```python
df["date_added"].fillna(0, inplace=True)
df["date_added"] = df["date_added"].astype(int)
df.head()
```

```python
# df[df["date_added"]==0]
df.loc[df["date_added"]==0, ]

# Check NaN values if exist or not?
df.info()

# check for duplicate rows
df.duplicated().sum()

# check for duplicate rows if any
df.duplicated().sum()

df.fillna("No Data", inplace=True)
df.isnull().sum()
```

```python
# Exploratory data analysis
plt.figure(figsize=(18, 8))
labels=['TV Show', 'Movie']
plt.pie(df['type'].value_counts().sort_values(),labels=labels,explode=[0.1,0.1],
        autopct='%1.5f%%',colors=['red','green'], startangle=90)
plt.title('Netflix Content Media Type')
plt.axis('equal')
plt.show()

# Conclusion
# There is much more to dig deep but these can be summarized as general inference.
```
